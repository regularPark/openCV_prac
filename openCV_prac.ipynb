```python
#cv2.imshow(image)
from google.colab.patches import cv2_imshow
cv2_imshow(image)
```

```python
image = cv2.resize(image, (640, 480))
cv2_imshow(image)
```

```python
image_gray = cv2.cvtColor(image, cv2.COLOR_RGB2GRAY)
cv2_imshow(image_gray)
```

```python
image.shape
image_gray.shape
```

```python
face_detector = cv2.CascadeClassifier('/content/drive/MyDrive/Colab Notebooks/cursos/haarcascade_frontalface_default.xml')
```

```python
detections = face_detector.detectMultiScale(image_gray)
```

```python
detections
```

```python
len(detections)
```

```python
for(x,y,w,h) in detections:
  #print(x,y,w,h)
  cv2.rectangle(image, (x,y), (x + w,y + h),(0,255, 0), 2)
cv2_imshow(image)
```
